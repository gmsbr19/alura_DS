```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dados = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
ultima_linha_laranja = 5000
coluna_diametro = 0
coluna_peso = 1

diametro_laranja = dados[:ultima_linha_laranja, coluna_diametro]
peso_laranja = dados[:ultima_linha_laranja, coluna_peso]

diametro_toranja = dados[ultima_linha_laranja:, coluna_diametro]
peso_toranja = dados[ultima_linha_laranja:, coluna_peso]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(peso_laranja, diametro_laranja)
plt.plot(peso_toranja, diametro_toranja)
plt.legend
```

# Regressão linear - Laranja

```python
X = peso_laranja
Y = diametro_laranja
n = peso_laranja.size

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)

plt.plot(X, Y)
plt.plot(X, X*a + b)
```

# Regresão linear - Toranja

```python
X = peso_toranja
Y = diametro_toranja
n = X.size

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)

plt.plot(X, Y)
plt.plot(X, X*a + b)
```
